```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("titanic.csv")
df
```

```python
golddf = pd.read_csv("gold price.csv")
golddf
```

```python
golddf["NewDate"] = [float("{}.{}".format(a.split("-")[0],a.split("-")[1])) for a in golddf["Date"]]
golddf
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
X = golddf["NewDate"]
y = golddf["Price"]

lr.fit(X.values.reshape(-1,1),y)
```

```python
predictPrice = lr.predict([[2020.08]])
predictPrice
```

```python
predicted = lr.predict(X.values.reshape(-1,1))
```

```python
plt.scatter(golddf["NewDate"],y,c="r")
plt.scatter(golddf["NewDate"],predicted,c="b")
plt.show()
```

```python
from sklearn.svm import SVR
```

```python
svr = SVR()
```

```python
svr.fit(X.values.reshape(-1,1),y)
```

```python
pridictPrice2 = svr.predict([[2020.08]])
pridictPrice2
```

```python
predicted2 = svr.predict(X.values.reshape(-1,1))
```

```python
plt.scatter(golddf["NewDate"],y,c='r')
plt.scatter(golddf["NewDate"],predicted2,c='b')
plt.show()
```
